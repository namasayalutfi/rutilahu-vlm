## Check File Formats in raw_img Folder

Analyze the file formats and their distribution in the raw_img directory.

```python
import os
from pathlib import Path
from collections import Counter

def check_file_formats(folder_path):
    """
    Check and display file formats in a given folder.
    
    Args:
        folder_path (str): Path to the folder to check
    """
    if not os.path.exists(folder_path):
        print(f"Folder tidak ditemukan: {folder_path}")
        return
    
    # Get all files in the folder
    files = os.listdir(folder_path)
    files = [f for f in files if os.path.isfile(os.path.join(folder_path, f))]
    
    if not files:
        print(f"Folder kosong: {folder_path}")
        return
    
    # Extract file extensions
    extensions = []
    for file in files:
        ext = Path(file).suffix.lower() if Path(file).suffix else 'no extension'
        extensions.append(ext)
    
    # Count extensions
    ext_counts = Counter(extensions)
    
    # Display results
    print(f"\n{'='*50}")
    print(f"File Formats Check: {folder_path}")
    print(f"{'='*50}")
    print(f"Total Files: {len(files)}\n")
    
    print("Format Breakdown:")
    print("-" * 50)
    for ext, count in sorted(ext_counts.items(), key=lambda x: x[1], reverse=True):
        percentage = (count / len(files)) * 100
        print(f"  {ext:15} : {count:3} files ({percentage:5.1f}%)")
    
    print("-" * 50)
    print("\nDetail Files:")
    print("-" * 50)
    for file in sorted(files):
        print(f"  {file}")
    print(f"{'='*50}\n")

# Check raw_img folder
raw_img_path = "../data/raw_img"
check_file_formats(raw_img_path)
```
